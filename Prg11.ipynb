```python
#11.Write a Python program to demonstrate Data Visualization using Seaborn
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
tips=pd.read_csv('/home/mcacnj/Dataset/tips.csv')
tips.head()
```

```python
# 1. Bar Plot - Total bill vs Day of the week
plt.figure(figsize=(8, 6))
sns.barplot(x="day", y="total_bill", data=tips, estimator=sum, ci=None, palette="Blues_d")
plt.title("Total Bill by Day of the Week")
plt.show()
```

```python
# 2. Scatter Plot - Total bill vs Tip with color coded by gender
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", hue="gender", style="gender", palette="deep", data=tips)
plt.title("Total Bill vs Tip (Gender Coded)")
plt.show()
```

```python
# 3. Box Plot - Total bill distribution by day of the week
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Total Bill Distribution by Day of the Week")
plt.show()
```

```python
# 4. Pair Plot - Pairwise relationships in the dataset
sns.pairplot(tips, hue="gender", palette="coolwarm")
plt.suptitle("Pairwise Relationships in the Tips Dataset", y=1.02)
plt.show()
```

```python
# 5. Heatmap - Correlation heatmap of numeric variables
corr = tips.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=1)
plt.title("Correlation Heatmap")
plt.show()
```

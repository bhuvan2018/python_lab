```python
#10.Write a python program to demonstrate Time series analysis with Pandas.
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt 
import seaborn as sns
df = pd.read_csv('/home/mcacnj/Dataset/AirPassengers.csv')
df.head(10)
```

```python
df.columns=['Month','Passengers']
df.head()
```

```python
def plot_df(df,x,y,title="",xlabel='Month',ylabel='Passengers',dpi=100):
    plt.figure(figsize=(15,4),dpi=dpi)
    plt.plot(x,y,color='tab:red')
    plt.gca().set(title=title,xlabel=xlabel,ylabel=ylabel)
    plt.show()
plot_df(df,x=df['Month'],y=df['Passengers'],title='Number of US Airline passengers from 1949 to 1950')
```

```python
x=df['Month'].values
y1=df['Passengers'].values
fig,ax=plt.subplots(1,1,figsize=(16,5),dpi=120)
plt.fill_between(x,y1=y1,y2=-y1,alpha=0.5,linewidth=2,color='seagreen')
plt.ylim(-800,800)
plt.title('Air Passengers (Two Side View)',fontsize=16)
plt.hlines(y=0,xmin=np.min(df['Month']),xmax=np.max(df['Month']),linewidth=.5)
plt.show()
```

```python
def plot_df(df,x,y,title="",xlabel='Month',ylabel='Passengers',dpi=100):
    plt.figure(figsize=(15,4),dpi=dpi)
    plt.plot(x,y,color='blue')
    plt.gca().set(title=title,xlabel=xlabel,ylabel=ylabel)
    plt.show()
plot_df(df,x=df['Month'],y=df['Passengers'],title='Trend and Seasonality')
```

```python
#1 line plot
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,4,6,8])
y=x*2
plt.plot(x,y,label="y=x*2",color="blue")
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#2
import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.rand(5,5),columns=['A','B','C','D','E'])
print(df)
df.plot.box(grid=True,vert=False,patch_artist=True,figsize=(8,6))
plt.title("Box plot sample")
plt.show()
```

```python
#4
import matplotlib.pyplot as plt
data=[18,21,22,19,20,25,30,35,28,30,40,50,60,70,25,24,35]
plt.hist(data,bins=5,edgecolor='black')
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#5
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=[[8,9,3,4],[4,6,9,1],[7,4,1,3],[9,5,1,3],[7,1,3,9]]
index=['I1','I2','I3','I4','I5']
columns=['C1','C2','C3','C4']
df=pd.DataFrame(data,index=index,columns=columns)
sns.heatmap(df,annot=True,cmap='coolwarm',cbar=True)
plt.title('Heatmap Example')
plt.show()
```

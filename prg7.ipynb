```python
import numpy as np
import matplotlib.pyplot as plt
def create_and_plot_array(m,n):
    array=np.random.randint(1,10,size=(m,n))
    print(f"Array: \n {array}\n")
    print(f"Shape: {array.shape}")
    print(f"Size: {array.size}")
    print(f"Datatype: {array.dtype}")
    plt.figure(figsize=(6,6))
    plt.imshow(array,cmap='viridis',interpolation='nearest')
    plt.colorbar(label='Value')
    plt.title('Heatmap of Array values')
    plt.show()
m,n=5,6
create_and_plot_array(m,n)
```

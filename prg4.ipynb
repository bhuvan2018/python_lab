```python
#1.) importing dataset
import pandas as pd
import numpy as np
data=pd.read_csv("file1.csv")
data.head(12)
```

```python
print(data.shape)
print(data.size)
```

```python
#2.) cleaning dataset
print("Before cleaning")
data.head(12)
```

```python
#remove duplicate rows
print("After cleaning")
data.drop_duplicates(subset=['qualification'], keep='last', inplace=True)
data.head(12)
```

```python
#after removing duplicate, total size in dataframes
print(data.shape)
print(data.size)
```

```python
#replacing existing values with new values
data.loc[7,'name']='sujatha'
data.head(12)
```

```python
#dropping the missing rows using dropna
df_dropped=data.dropna(axis=0,how="all",inplace=True)
data.head(10)
```

```python
mean_val=data['mobnumber'].mean()
data['mobnumber']=data['mobnumber'].fillna(mean_val)
data.head(10)
```

```python
data['city']=data['city'].fillna('kanyana')
data.head(8)
```

```python
data.dropna(subset=['city'],inplace=True)
data.head(10)
```

```python
data['qualification']=data['qualification'].fillna('Msc')
data.head(12)
```
